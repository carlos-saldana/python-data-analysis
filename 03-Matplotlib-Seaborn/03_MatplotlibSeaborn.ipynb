#### MÉTODO ORIENTADO A OBJETOS DE MATPLOTLIB

```python
#Importamos nuestras librerías
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creación de variables
x = np.linspace(0, 5, 11)
y = x**2
```

```python
#Utilizamos el concepto de objeto para crear una gráfica más personalizable
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.5])
axes.plot(x, y, "b")
plt.show()
```

```python
#Podemos crear varias gráficas dentro del lienzo
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.17, 0.55, 0.4, 0.3])

axes.plot(x, y, "b:")
axes2.plot(y, x, "black")
axes2.set_facecolor("green")

plt.show()
```

#### SUBPLOTS

```python
#Importamos las librerías a emplear
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creamos nuestras variables
x = np.linspace(0, 5, 11)
y = x**2
```

```python
#Creacmos un gráfico
fig, axes = plt.subplots()
axes.plot(x, y)
plt.show()
```

```python
#Creamos varias gráficas tipo fig
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, "b")
axes[0].set_title("Gráfica 1")
axes[0].set_xlabel("Eje x")
axes[0].set_ylabel("Eje y")

axes[1].plot(y, x, "r")
axes[1].set_title("Gráfica 2")
axes[1].set_xlabel("Eje x2")
axes[1].set_ylabel("Eje y2")

plt.show()
```

```python
#Accedemos a los gráficos mediante el nombre almacenado en una tupla
fig, (axes1, axes2) = plt.subplots(nrows=1, ncols=2)

axes1.plot(x, y, "r")
axes2.plot(y, x, "b")

plt.show()
```

```python
#Generamos una matriz de tipo fig, con 2 filas y 4 columnas
fig, axes = plt.subplots(2, 4)

axes[0,0].plot(x, y)
axes[0,1].plot(y, x, "r")

#Mejora la visualización de los ejes 
fig.tight_layout()
```

```python
#También podemos trabajar mediante tuplas
fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)

axes1.plot(x, np.sin(x), 'b')
axes2.plot(y, np.cos(x), 'purple')
axes3.plot(x, np.tan(y), 'orange')
axes4.plot(y, np.cos(y)**2,'black')

#Mejora la visualización de los ejes 
fig.tight_layout()
```

#### EJEMPLO INTERESANTE: ARMÓNICO

```python
#Creación de variables
x = np.linspace(0, 20, 100)
y = np.zeros(100)
n = np.linspace(1, 9, 9, endpoint=True)
harmonics = []
```

```python
#Función for
for idx, element in enumerate(n):
    y = np.sin(x*element)
    harmonics.append(y)
    
harmonics = np.array(harmonics)
harmonics = np.expand_dims(harmonics, axis=0)
harmonics = harmonics.reshape((3, 3, 100))
```

```python
#Gráfica
nrow = 3
ncol = 3

fig, axes = plt.subplots(nrow, ncol)

for idx in range(nrow):
    for jdx in range(ncol):
        axes[idx, jdx].plot(x, harmonics[idx, jdx])
        axes[idx, jdx].set_title(f"Harmonic of sin(x)")
        
fig.tight_layout()
```

#### BAR PLOT
Gráfico de barras que nos permite visualizar variables categóricas.

```python
#Importamos las librerías las librerías
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Variables categóricas
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", "GERMANY"]
population = [5000, 800, 900, 1000, 300]
```

```python
#Gráfica de barras
plt.bar(countrys, population)
plt.show()
```

```python
#Modificamos colores y tamaño
plt.bar(countrys, population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.show()
```

```python
#Modificar el eje x
plt.bar(countrys, population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

```python
#Gráfico de barras horizontal
plt.barh(countrys, population)
plt.show()
```

#### OTROS TIPOS DE GRÁFICOS

```python
#Empezamos creando nuestra data
data = np.random.randint(1, 50, 100)
```

```python
#Histograma
plt.hist(data)
plt.show()
```

```python
#Número de barras con Bins
plt.hist(data, bins = 50)
plt.show()
```

```python
#Cambiamos el estilo
plt.hist(data, bins=10, histtype="step")
plt.show()
```

```python
#Gráfico tipo Boxplot
plt.boxplot(data)
plt.show()
```

```python
#Modificaciones
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
#Quitar outliers
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)  
plt.show()
```

```python
#Gráfico Scatter

#Declaramos nuestras variables
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)

#Graficamos
plt.scatter(x,y, s=area)
plt.show()
```

```python
#Cambiamos el color y forma
plt.scatter(x, y, s=area, c=colors, marker='p')
plt.show()
```

```python
#Transparencia para ver los superpuestos
plt.scatter(x, y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
