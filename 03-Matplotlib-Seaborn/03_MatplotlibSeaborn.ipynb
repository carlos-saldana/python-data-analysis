#### SUBPLOTS

```python
#Importamos las librerías a emplear
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creamos nuestras variables
x = np.linspace(0, 5, 11)
y = x**2
```

```python
#Creacmos un gráfico
fig, axes = plt.subplots()
axes.plot(x, y)
plt.show()
```

```python
#Creamos varias gráficas tipo fig
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, "b")
axes[0].set_title("Gráfica 1")
axes[0].set_xlabel("Eje x")
axes[0].set_ylabel("Eje y")

axes[1].plot(y, x, "r")
axes[1].set_title("Gráfica 2")
axes[1].set_xlabel("Eje x2")
axes[1].set_ylabel("Eje y2")

plt.show()
```

```python
#Accedemos a los gráficos mediante el nombre almacenado en una tupla
fig, (axes1, axes2) = plt.subplots(nrows=1, ncols=2)

axes1.plot(x, y, "r")
axes2.plot(y, x, "b")

plt.show()
```

```python
#Generamos una matriz de tipo fig, con 2 filas y 4 columnas
fig, axes = plt.subplots(2, 4)

axes[0,0].plot(x, y)
axes[0,1].plot(y, x, "r")

#Mejora la visualización de los ejes 
fig.tight_layout()
```

```python
#También podemos trabajar mediante tuplas
fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)

axes1.plot(x, np.sin(x), 'b')
axes2.plot(y, np.cos(x), 'purple')
axes3.plot(x, np.tan(y), 'orange')
axes4.plot(y, np.cos(y)**2,'black')

#Mejora la visualización de los ejes 
fig.tight_layout()
```

#### EJEMPLO INTERESANTE: ARMÓNICO

```python
#Creación de variables
x = np.linspace(0, 20, 100)
y = np.zeros(100)
n = np.linspace(1, 9, 9, endpoint=True)
harmonics = []
```

```python
#Función for
for idx, element in enumerate(n):
    y = np.sin(x*element)
    harmonics.append(y)
    
harmonics = np.array(harmonics)
harmonics = np.expand_dims(harmonics, axis=0)
harmonics = harmonics.reshape((3, 3, 100))
```

```python
#Gráfica
nrow = 3
ncol = 3

fig, axes = plt.subplots(nrow, ncol)

for idx in range(nrow):
    for jdx in range(ncol):
        axes[idx, jdx].plot(x, harmonics[idx, jdx])
        axes[idx, jdx].set_title(f"Harmonic of sin(x)")
        
fig.tight_layout()
```

#### BAR PLOT
Gráfico de barras que nos permite visualizar variables categóricas.

```python
#Importamos las librerías las librerías
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Variables categóricas
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", "GERMANY"]
population = [5000, 800, 900, 1000, 300]
```

```python
#Gráfica de barras
plt.bar(countrys, population)
plt.show()
```

```python
#Modificamos colores y tamaño
plt.bar(countrys, population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.show()
```

```python
#Modificar el eje x
plt.bar(countrys, population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

```python
#Gráfico de barras horizontal
plt.barh(countrys, population)
plt.show()
```

### SEABORNS
Librería construida sobre Matplotlib e integrada para estructuras de Pandas, optimizada para funcionar con DataFrames.
Ventajas:
- Gran velocidad.
- Facilidad para escribir código.
- Altamente customizable entre gráficas y visualizaciones.

```python
#Importamos las librerías a emplear
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#GRÁFICA DE BARRAS

#Declaramos nuestros ejes
a = ['A', 'B', 'C']
b = [1, 3, 2]

#Graficamos
sns.barplot(x=a, y=b)
plt.show()
```

```python
#Configuramos fuentes y estilos
sns.set(style = "dark", palette = "dark", font ="Verdana", font_scale = 0.8)
sns.barplot(x = a, y = b)
plt.show()


""""
NOTAS:
- El método SET afecta a TODAS las gráficas.
- Podemos resetar los valores utilizando el método reset_orig.
- Este parámetro es un atajo de set_theme.
"""
```

```python
#Parámetros más comunes
tips = sns.load_dataset("tips")
tips
```

```python
#Gráfico de frecuencia para el TOTAL_BILL
sns.displot(data = tips, x = "total_bill");
```

```python
#Agregamos una variable Y & seaborn adapta la gráfica
sns.displot(data = tips, x = "total_bill", y = "tip")
plt.show()
```

```python
#Hacemos agrupamiento por una variable (hue)
sns.displot(data = tips, x = "total_bill", y = "tip", hue = "sex");
```

```python
#Cambiamos el tipo de gráfica
sns.displot(data = tips, x = "total_bill", hue = "sex", kind = "kde")
plt.show()
```

```python
#Quitamos la leyenda, cambiamos la paleta de colores y transparencia
sns.displot(data = tips, x = 'total_bill', hue = 'sex', kind = 'kde', legend= False, palette='dark', alpha = .5);
```

#### GRÁFICOS DE RELACIÓN (SIMILARES A SCATTER EN MATPLOTLIB)

```python
#Importamos las librerías a emplear
import seaborn as sns
import matplotlib.pyplot as plt 

#Cargamos nuestro dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#SCATTER
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day');
```

```python
#Cambiamos el estilo de los puntos y tamaño
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
```

```python
#Cambiamos el formato de los marcadores
markers = {"Lunch": 'D', "Dinner": 's'}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers);
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5));
```

```python
#Gráfica de líneas
sns.lineplot(data=tips, x='total_bill', y='tip',  hue='day', style='time', size='size');
```

```python
#REPLOT: Con KIND podemos modificar el tipo de gráfico
sns.relplot(data=tips, x='total_bill', y='tip',  hue='day', style='time', size='size', kind='line');
```

```python
#Agregamos el formato COL
sns.relplot(data=tips, x='total_bill', y='tip',  hue='day', style='time', size='size', kind='scatter', col='time');
```

#### JOINTPLOT AND PAIRPLOT

```python
#Cargamos nuestro dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#JOINPLOT: Unimos 2 distintas gráficas en una sola
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex');
```

```python
#Agregamos más argumentos para poder hacer un mejor análisis
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, #marginal_ticks: Tabla para la gráfica exterior
marginal_kws = dict(bins=25, fill=True, multiple='dodge')); #marginal_kws: Se le pueden agregar diferentes parámetros
```

```python
#PAIRPLOT: Correlaciona todas las variables numéricas que hay
sns.pairplot(data = tips)
plt.show()
```

```python
#Solo mostramos una esquina del PAIRPLOT, eliminando las gráficas repetidas
sns.pairplot(data = tips, corner = True);
```
